# <font face="黑体" color="purple"> 向量</font>

## <font face="黑体" color=DarkViolet> 向量的概念</font>

n维向量指n个数构成的有序数组.
比如,列向量:$(a1,a2..….,an)^T$，$a_i$称为向量的第i个分量，如果向量所有的分量都是0，则称为零向量$0=(0,0...o)^T$

向量空间中的每个向量除$0$向量外都有一个长度，长度的术语是**范数**，$\mathbb{R}^k$中向量的范数的一般公式是
$$
\|u\|= \sqrt{x_1^2+x_2^2+...+x_k^2}
$$

下列例子将在二维空间中绘出向量 $(4, 7)$ 和 $(8, 6)$ 的范数

```python
fig, ax = plt.subplots(figsize=(8, 8))
vec = np.array([[0, 0, 4, 7], [0, 0, 8, 6]])

X, Y, U, V = zip(*vec)
ax.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1, color = 'red', alpha = .6)
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_xlabel('x-axis', fontsize =16)
ax.set_ylabel('y-axis', fontsize =16)
ax.grid()
ax.text(4, 7, '$(4, 7)$', fontsize = 16)
ax.text(8, 6, '$(8, 6)$', fontsize = 16)

ax.plot([4, 4], [0, 7], c = 'b', lw = 2)
ax.plot([0, 4], [0, 0], c= 'b', lw = 2)

ax.plot([8, 0], [0, 0], c = 'b', lw = 2)
ax.plot([8, 8], [0,6], c= 'b', lw = 2)

ax.text(1.7, 3.8, '$\sqrt{4^2+7^2}$', fontsize = 16, rotation = np.arctan(7/4)*180/np.pi)
ax.text(5, 4.1, '$\sqrt{8^2+6^2}$', fontsize = 16, rotation = np.arctan(6/8)*180/np.pi ) 
# use arctan to get the rotation degree and convert from radian to degrees
plt.show()
```

## <font face="黑体" color=DarkViolet> 向量的运算</font>

向量加法是元素级操作, 向量 $\mathbf{u}$ 和 $\mathbf{v}$ 相加如下所示:

$$\mathbf{u}+\mathbf{v}=\left[\begin{array}{c}
u_{1} \\
u_{2} \\
\vdots \\
u_{n}
\end{array}\right]+\left[\begin{array}{c}
v_{1} \\
v_{2} \\
\vdots \\
v_{n}
\end{array}\right]=\left[\begin{array}{c}
u_{1}+v_{1} \\
u_{2}+v_{2} \\
\vdots \\
u_{n}+v_{n}
\end{array}\right]$$

减法:
$$\mathbf{u}+(-\mathbf{v})=\left[\begin{array}{c}
u_{1} \\
u_{2} \\
\vdots \\
u_{n}
\end{array}\right]+\left[\begin{array}{c}
-v_{1} \\
-v_{2} \\
\vdots \\
-v_{n}
\end{array}\right]=\left[\begin{array}{c}
u_{1}-v_{1} \\
u_{2}-v_{2} \\
\vdots \\
u_{n}-v_{n}
\end{array}\right]$$

标量乘法:
$$c \mathbf{u}=c\left[\begin{array}{c}
u_{1} \\
u_{2} \\
\vdots \\
u_{n}
\end{array}\right]=\left[\begin{array}{c}
c u_{1} \\
c u_{2} \\
\vdots \\
c u_{n}
\end{array}\right]$$

在二维空间和三维空间展示向量加法

```python
fig, ax = plt.subplots(figsize = (9,9))

vec = np.array([[[0, 0, 4, 7]],
               [[0, 0, 8, 4]],
               [[0, 0, 12, 11]],
               [[4, 7, 8, 4]],
               [[8, 4, 4, 7]]])
color = ['r','b','g','b','r']

for i in range(vec.shape[0]):
    X,Y,U,V = zip(*vec[i,:,:])
    ax.quiver(X, Y, U, V, angles='xy', scale_units='xy', color = color[i], scale=1, alpha = .6)

ax.set_xlim([0, 15])
ax.set_ylim([0, 15])
ax.set_xlabel('x-axis', fontsize =16)
ax.set_ylabel('y-axis', fontsize =16)
ax.grid()

for i in range(3):
    ax.text(x = vec[i,0,2], y = vec[i,0,3], s = '(%.0d, %.0d)' %(vec[i,0,2],vec[i,0,3]), fontsize = 16)

ax.text(x= vec[0,0,2]/2, y = vec[0,0,3]/2, s= '$u$', fontsize = 16)
ax.text(x= 8, y = 9, s= '$v$', fontsize = 16)
ax.text(x= 6, y = 5.5, s= '$u+v$', fontsize = 16)

ax.set_title('向量加法', size = 18)
plt.show()
```

## <font face="黑体" color=DarkViolet> 线性组合 </font>

m个n维向量$a_1$ .... $a_m$以及m个实数,
把这种结构$k_1a_1+ ... +k_ma_m$称为$a_1$ ... $a_m$的一个线性组合$k_1$ ... $k_m$为线性组合的系数



## <font face="黑体" color=DarkViolet> 线性表出 </font>
一个向量β如果能够表示为$k_1a_1+...+k_ma_m$的线性组合，存在一组实数$k_1... k_m$使得$β= k_1a_1...+k_ma_m$，称向量β可以由$a_1...a_m$线性表出


考虑 $\mathbb{R}^2$ 中的两个向量 $u$ 和 $v$ ，它们相互独立，即不指向相同或相反的方向。因此，$\mathbb{R}^k$ 中的任何向量都可以由 $u$ 和 $v$ 的**线性组合**表示。

例如，这是一个线性组合本质上是一个线性系统。


$$
c_1
\left[
\begin{matrix}
4\\
2
\end{matrix}
\right]+
c_2
\left[
\begin{matrix}
-2\\
2
\end{matrix}
\right]
=
\left[
\begin{matrix}
2\\
10
\end{matrix}
\right]
$$

使用SymPy解出此系统

```python
A = sy.Matrix([[4, -2, 2], [2,2,10]])
A.rref()
```

结果是 $(c_1, c_2)^T = (2, 3)^T$，指 $c_1$ 和 $c_2$分别等于2和3

在该系统中绘制由$u$和$v$为基张成的空间

计算向量的斜率, i.e. $\frac{y}{x}$
$$
s_1 =\frac{y}{x} = \frac{2}{4}=0.5\\
s_2 =\frac{y}{x}= \frac{2}{-2}=-1
$$

基可以如下构建：
$$
y_1 = a+0.5x\\
y_2 = b-x
$$
其中$a$和$b$将被设为等间隔的常数, 例如$(2.5, 5, 7.5, 10)$. 
坐标的基是粉红色的的网格，每个线段是一个单位(如‘1’在笛卡尔坐标系中)在“新”坐标。

```python
fig, ax = plt.subplots(figsize=(8, 8))

vec = np.array([[[0,0,4,2]],
                 [[0,0,-2,2]],
                 [[0,0,2,10]],
                 [[0,0,8,4]], 
                 [[0,0,-6,6]]])
colors = ['b','b','r','b','b']

for i in range(vec.shape[0]):
    X,Y,U,V = zip(*vec[i,:,:])
    ax.quiver(X, Y, U, V, angles='xy', scale_units='xy', color = colors[i], scale=1, alpha = .6)
    ax.text(x = vec[i,0,2], y = vec[i,0,3], s = '(%.0d, %.0d)' %(vec[i,0,2],vec[i,0,3]), fontsize = 16)

points12 = np.array([[8,4],[2,10]])
ax.plot(points12[:,0], points12[:,1], c = 'b', lw = 3.5,alpha =0.5, ls = '--')

points34 = np.array([[-6, 6],[2,10]])
ax.plot(points34[:,0], points34[:,1], c = 'b', lw = 3.5,alpha =0.5, ls = '--')

ax.set_xlim([-10, 10])
ax.set_ylim([0, 10.5])
ax.set_xlabel('x-axis', fontsize =16)
ax.set_ylabel('y-axis', fontsize =16)
ax.grid()


a = np.arange(-11, 20, 1)
x = np.arange(-11, 20, 1)

for i in a:    
    y1 = i + .5*x
    ax.plot(x, y1, ls = '--', color = 'pink', lw = 2)
    y2 = i - x
    ax.plot(x, y2, ls = '--', color = 'pink', lw = 2)    

ax.set_title('二维空间中两向量的线性组合', size = 22, x =0.5, y = 1.01)
plt.show()
```

## <font face="黑体" color=DarkViolet> 线性相关 </font>

若有m个n维向量$a_1...a_m$,存在**不全为零**的实数$k_1...k_m$，使得$k_1a_1+…+k_ma_m$ = 0成立,则称向量组$a_1...a_m$线性相关，否则称其线性无关


重要结论:<br>
1.如果向量组中有零向量，那这个向量小组肯定线性相关，线性相关的一个充分条件<br>
2.如果向量小组中有坐标成比例的两个向量，那这个向量小组肯定线性相关，线性相关的一个充分条件<br>
3.如果两个向量的坐标不成比例，那这两向量就是线性无关<br>
4.n个方程n个未知数判断有没有非零解,才可以用行列式


## <font face="黑体" color=DarkViolet> 线性无关 </font>

由线性相关的定义：<br>
m个n维向量$a_1...a_m$,不存在不全为零的实数$k_1...k_m$，使得$k_1a_1+…+k_ma_m$ = 0成立，则称称其线性无关




判断${v}_1, {v}_2, {v}_3$ 是否线性相关,检验的常用方法是构造增广矩阵并计算其简化形式

例如：<br>
$$
{v}_{1}=\left[\begin{array}{l}
1 \\
2 \\
3
\end{array}\right]^T, 
{v}_{2}=\left[\begin{array}{l}
4 \\
5 \\
6
\end{array}\right]^T, \text { and } {v}_{3}=\left[\begin{array}{l}
2 \\
1 \\
0
\end{array}\right]^T
$$

```python
A = sy.Matrix([[1,4,2],
               [2,5,1],
               [3,6,0]])
A.rref()
```

从解可以看出，$x_3$是一个自由变量，所以它自然可以是非零的，因为$x_3\cdot 0 =0$，因此这组向量是线性相关的。

再例如<br>
矩阵 $A$ 如下，判断其各列是否线性无关

$$
A=\left[\begin{array}{rrr}
0 & 1 & 4 \\
1 & 2 & -1 \\
5 & 8 & 0
\end{array}\right]
$$

```python
A = sy.Matrix([[0,1,4,0],[1,2,-1,0],[5,8,0,0]])
A.rref()
```

此时方程有3个基本变量，没有自由变量，$A$的各列是线性无关的。


## <font face="黑体" color=DarkViolet> 线性无关的可视化 </font>

可视化展示一组线性相关向量$(3, 2)^T$, $(-9, -6)^T$, $(6, 4)^T$

```python
fig, ax = plt.subplots(figsize = (8, 8))
 
arrows = np.array([[[0,0,3,2]],
                   [[0,0,-9,-6]], 
                   [[0,0,6,4]]])
colors = ['r','b','g']
for i in range(arrows.shape[0]):
    X,Y,U,V = zip(*arrows[i,:,:])
    ax.arrow(X[0], Y[0], U[0],V[0], color = colors[i], width = .18, 
             length_includes_head = True,
             head_width = .3, # default: 3*width
             head_length = .6,
             overhang = .4, zorder = -i)

ax.scatter(0, 0, ec = 'red', fc = 'black', zorder = 5)
ax.text(6, 4, '$(6, 4)$')
ax.text(-9, -6.5, '$(-9, -6)$')
ax.text(3, 2, '$(3, 2)$')

ax.grid(True)
ax.set_title('线性相关可视化')
ax.axis([-10, 10, -10, 10])
ax.set_xlabel('x-axis', size = 18)
ax.set_ylabel('y-axis', size = 18)
plt.show()
```
